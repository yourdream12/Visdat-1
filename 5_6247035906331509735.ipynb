```python
import pandas as pd
import bokeh.io
from bokeh.io import output_notebook
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource, NumeralTickFormatter
from bokeh.models import HoverTool
from bokeh.models.widgets import Tabs, Panel
from bokeh.models import ColumnDataSource, CDSView, GroupFilter
```

```python
df= pd.read_csv("AGRO.csv")
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
df = df.rename(columns={"Adj Close" : "Adj_Close"})
df.head()
```

```python
df.isnull().sum()
```

```python
# drop kolom data
columns = ['first_trade','close','value','frequency','index_individual','offer','offer_volume','bid','bid_volume','listed_shares','tradeble_shares','weight_for_index','foreign_sell','foreign_buy','delisting_date','non_regular_volume','non_regular_value','non_regular_frequency']
df.drop(columns, inplace=True, axis=1, errors='ignore')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# Output notebook
output_notebook()

# Store the data in a ColumnDataSource
df_cds = ColumnDataSource(df)

# Specify the selection tools to be made available
select_tools = ['box_select', 'lasso_select', 'poly_select', 'tap', 'reset']

# Create the figure
fig = figure(plot_height=600,
             plot_width=800,
             x_axis_type='datetime',
             x_axis_label='date',
             y_axis_label='volume',
             title='Visualisasi pasar saham AGRO',
             toolbar_location='below',
             tools=select_tools)


# Add square representing each data
fig.square(x='date',
           y='volume',
           source=df_cds,
           color='royalblue',
           selection_color='deepskyblue',
           nonselection_color='lightgray',
           nonselection_alpha=0.3)

# Configure a renderer to be used upon hover
hover_glyph = fig.circle(x='date', y='volume', source=df_cds,
                         size=15, alpha=0,
                         hover_fill_color='black', hover_alpha=0.5)


# Format the tooltip
tooltips = [
            ('date','@Date'),
            ('previous', '@previous'),
            ('open_price', '@open_price'),
            ('high', '@high'),
            ('low', '@low'),
            ('changee', '@change'),
            ('volume', '@volume'),
           ]

# Add the HoverTool to the figure
fig.add_tools(HoverTool(tooltips=tooltips, renderers=[hover_glyph]))

# Visualize
show(fig)
```

```python
output_notebook()
output_file('level2.html',
            title='Level 2')
select_tools = ['pan','box_select','tap', 'reset','save']
data = ColumnDataSource(df)
high_fig = figure(plot_height=800, x_axis_type='datetime', plot_width=800, x_axis_label='date', y_axis_label='high', title='Visualisasi pasar saham AGRO', toolbar_location='right', tools=select_tools)

low_fig = figure(plot_height=800, x_axis_type='datetime', plot_width=800, x_axis_label='date', y_axis_label='low', title='Visualisasi pasar saham AGRO', toolbar_location='right', tools=select_tools)

volume_fig = figure(plot_height=800, x_axis_type='datetime', plot_width=800, x_axis_label='date', y_axis_label='volume', title='Visualisasi pasar saham AGRO', toolbar_location='right', tools=select_tools)

high_fig.circle(x='date', y='high', source=data, color='magenta', selection_color='k', nonselection_color='mediumpurple', nonselection_alpha=0.4)

low_fig.circle(x='date', y='low', source=data, color='magenta', selection_color='k', nonselection_color='g', nonselection_alpha=0.4)

volume_fig.circle(x='date', y='volume', source=data, color='magenta', selection_color='k', nonselection_color='r', nonselection_alpha=0.4)

hover_glyph1 = high_fig.circle(x='date', y='high', source=data, size=15, alpha=0, hover_fill_color='navy', hover_alpha=0.7)

hover_glyph2 = low_fig.circle(x='date', y='low', source=data, size=15, alpha=0, hover_fill_color='green', hover_alpha=0.7)

hover_glyph3 = volume_fig.circle(x='date', y='volume', source=data, size=15, alpha=0, hover_fill_color='crimson', hover_alpha=0.7)

tooltips =[('date ','@date'),('volume ','@volume'),('open_price ','@open_price'),('high ','@high'),('low ','@low')]

high_fig.add_tools(HoverTool(tooltips=tooltips, renderers=[hover_glyph1]))
low_fig.add_tools(HoverTool(tooltips=tooltips, renderers=[hover_glyph2]))
volume_fig.add_tools(HoverTool(tooltips=tooltips, renderers=[hover_glyph3]))

high_panel1 = Panel(child=high_fig, title='high')
low_panel2 = Panel(child=low_fig, title='low')
volume_panel3 = Panel(child=volume_fig, title='volume')

tabs = Tabs(tabs=[high_panel1,low_panel2,volume_panel3])

# show(fig)
show(tabs)
```
